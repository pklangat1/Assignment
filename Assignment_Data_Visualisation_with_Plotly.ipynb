```python
# Question: Create a pie chart to describe the distribution of olympic gold medals.
# ---
# Dataset URL = http://bit.ly/OlympicsDataset
# ---
# 

# Explanation
# ---
# `color_discrete_sequence` argument sets the color of pie sectors.  
#  For more colors visit https://plotly.com/python/builtin-colorscales/ 
# `hole` creates a donut chart.


# Prepare
olympics_df = pd.read_csv('http://bit.ly/OlympicsDataset')
olympics_df.head()
```

```python
# Render
fig = px.pie(
    olympics_df, 
    values  = 'gold_medal', 
    names   = 'country',
    color_discrete_sequence=px.colors.sequential.RdBu,
    hole = 0.5
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': 'Olympic Gold Medal Distribution', 
        'x': 0.45,
        'xanchor': 'center',
        'yanchor': 'top'}
  )
fig.update_traces(textposition='inside', textinfo='percent+label')

# Display
fig.show()
```

```python
# Question
# ---
# Create a histogram to display the frequency distribution of the overall variable in the FIFA dataset.
# ---
# Dataset URL = http://bit.ly/FifaDS
# ---
# 
# YOUR CODE GOES BELOW
# 

# Explanations
# ---
# `histogram()` function creates a histogram from a variable in the dataset 
#  with the argument `nbins` defining the number of bins.
# `bargap` parameter in `update_layout()` defines the gap between the histogram bars.

# Prepare
fifa_df = pd.read_csv('http://bit.ly/FifaDS')
fifa_df.head()
```

```python
# Render
fig = px.histogram(
    fifa_df, 
    x = "Overall",
    nbins = 10
  )

# Tweak and Label
fig.update_layout(
    title = {
        'text': 'Histogram of Overall Player Rating ', 
        'x': 0.45,
        'xanchor': 'center',
        'yanchor': 'top'},
    bargap = 0.1,
    xaxis_title_text ='Overall Rating',
    yaxis_title_text ='Number of Players',
  ) 

# Display
fig.show()
```

```python
# Example 2
# ---
# Create a line graph to describe the trend of gas price/gallon in the UK, USA and France since 1990.
# ---
# 

# Explanation
# ---
# Inside `line()` we map UK, USA and France variables on the y axis using a list.
# `labels` argument sets the name of the labels.
# `template` argument  sets the theme. More themes: https://plotly.com/python/templates/

# We first convert dataframe into long format for use in `line()`
# Further reading: https://bit.ly/3myLrWg 
gas_df2 = pd.melt(gas_df, id_vars=['Year'], value_vars=['UK', 'USA', 'France'])
gas_df2.head()
```
